```python
def give_recommendations(user_series, dframe, series_genredf):
  inputSeries = pd.DataFrame(user_series)
  seriesId = dframe[dframe['Series_Title'].isin(inputSeries['Series_Title'].tolist())]
  inputSeries = pd.merge(seriesId, inputSeries)
  inputSeries = inputSeries.drop('Genre', 1)
  user_series = series_genredf[series_genredf['Series_Title'].isin(inputSeries['Series_Title'].tolist())]
  user_series = user_series.reset_index(drop=True)
  print(inputSeries)
  userGenreTableSer = user_series.drop('Series_Title', 1)
  userProfileSer = userGenreTableSer.transpose().dot(inputSeries['rating'])
  #print(userProfileSer)
  genreTableSer = series_genredf.set_index(series_genredf['Series_Title'])
  genreTableSer = genreTableSer.drop('Series_Title', 1)
  recommendationTable_dfSer = ((genreTableSer*userProfileSer).sum(axis=1))/(userProfileSer.sum())
  recommendationTable_dfSer = recommendationTable_dfSer.sort_values(ascending=False)
  return dframe.loc[dframe['Series_Title'].isin(recommendationTable_dfSer.head(20).keys())]
```
